```python
import numpy as np
IMG_SIZE = (224,224)


m1 = tf.keras.models.load_model("../outputs/model/densenet121_model.h5", compile=False)
m2 = tf.keras.models.load_model("../outputs/model/densenet121_model.keras", compile=False)

def load_image(path):
    img = tf.keras.utils.load_img(
        path,
        target_size=IMG_SIZE
    )
    img = tf.keras.utils.img_to_array(img)
    img = img / 255.0
    img = np.expand_dims(img, axis=0)  # (1, 224, 224, 3)
    return img


x = load_image('glioma_test.jpg')

print(m1.predict(x))
print(m2.predict(x))
```
